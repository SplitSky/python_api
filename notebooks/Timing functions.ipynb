```python
def getChiSqrt(fit_y, y, ey):
    # all arrays are numpy arrays
    # returns the chi squared value
    chi_sqrt = ((y - fit_y) / (ey)) ** 2
    return np.sum(chi_sqrt)

def print_coeff(coff, error):
    print("Printing coeffcients")
    print('constant term = {:.3g} +/- {:.2g}'.format(coff[1],error[1][1]))
    print('linear term   = {:.3g} +/- {:.2g}'.format(coff[0],error[0][0]))
    print(" ")
    

    
def plot_Data(x, y, error_y, label, color):
    y_weights = (1 / error_y) * np.ones(np.size(y))
    y_errors = error_y * np.ones(np.size(y))
    fit_parameters, fit_errors = np.polyfit(x, y, 1, cov=True, w=y_weights)
    y_fitted = np.polyval(fit_parameters, x)
        
    print(f'Coefficients for {label}')
    temp = getChiSqrt(y_fitted, y, y_errors)
    print(f'Chi squared: {temp}')
    print_coeff(fit_parameters,fit_errors)   
    plt.scatter(x,y,color=color,marker='+',label=label)
    plt.plot(x, y_fitted, color=color)
    plt.errorbar(x,y,error_y,color=color, linestyle=' ')
    plt.legend()
    
def get_errors(x,y):
    coff = np.polyfit(x,y,3)
    x = np.array(x)
    y = np.array(y)
    f = lambda x, coff : coff[3] + coff[2]*x + coff[1]*x**2 + coff[0]*x**3
    
    std = np.std(y-f(x,coff))
    #return np.array(x)*0 + np.mean(np.abs(y-f(x, coff)))
    return np.array(x)*0 + std*2
    
import numpy as np
import matplotlib.pyplot as plt
# old data
x_abs = np.arange(0,100000,1000)
x=[0, 10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000]
y=[1.9370907610000359, 14.31611829800022, 27.76208614299958, 43.01399246499932, 57.06589911500032, 69.40096504699977, 83.23451032699995, 97.79610589299955, 113.03963722800017, 128.0035330820001]


ey = get_errors(x,y)
plot_Data(x,y,ey, "Semester 1","b")

# number of dataset entries dependency on time
# new data - no cache or no multithread



x = [0, 10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000]
y = [1.59718059899933, 12.040914136000538, 22.20516422899982, 33.275151013999675, 42.286958357, 51.80988918799994, 62.6146269840001, 72.28213849299937, 82.86467851399993, 93.00410126500083]


# new data with cache enabled
x= [0, 10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000]
y=[0.8611950840004283, 10.754028700999697, 21.363570642999548, 31.389242595000724, 41.91990103900025, 51.21813546300018, 62.073974048999844, 71.68615343599959, 82.32624440200016, 92.2875103410006]

ey = get_errors(x,y)/2
plot_Data(x,y,ey, "Semester 2 with cache","g")


# new data with cache and multithreading
x= [0, 10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000]
y=[0.8867142600001898, 10.651210166000055, 21.16087392300051, 31.30209599799946, 42.02718274400013, 51.761872557999595, 62.87305915900015, 72.11619896100001, 82.08008494799924, 92.2495490429992]
ey = get_errors(x,y)/2
plot_Data(x,y,ey, "Semester 2 cache and multithread","r")


plt.savefig("dat_size_vs_time.png")
plt.title("Time complexity - Dataset size against time")
plt.xlabel("Number of dataset entries N")
plt.ylabel("Time taken in seconds")
plt.legend()
plt.show()
plt.savefig("dat_size_vs_time.png")
```
